```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state =42)
```

```python
from sklearn.cluster import KMeans
```

```python
# manual process 
# elbow method to select the k value

wcss = []
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
# warnings due to mismatch in version
wcss
# here wcss is declining which is very good that we wanted
```

```python
# to draw the elbow graph
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.ylabel('WCSS')
plt.xlabel('No of clusters')
# plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++') # here we have found that the at the 3 the were the wcss getting normal 
```

```python
y_labels=kmeans.fit_predict(x_train) #bc here we have to find the labels of our x_train data's usually refers to the output or target variable that the model is trained to predict based on input features  
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_labels)
```
